```python
file_names = ['Abortion', 'Euthanasia', 'Gene_Editing', 'Human_Clinical_Trials', 'Mandatory_Vaccination', 'Organ_Transplant']

avg_list = []

for item in file_names:
    df = pd.read_csv(f'all_data/{item}.csv')
    word_counts = df['text'].str.split().str.len()
    average_word_count = word_counts.mean()
    avg_list.append(average_word_count)
    print(f"{item}: {round(average_word_count, 1)}, len(df): {len(df)}")

overall_avg = round(sum(avg_list) / len(avg_list), 1)
print(f"所有话题的平均单词数: {overall_avg}")
```

```python
import pandas as pd

file_names = ['Abortion', 'Euthanasia', 'Gene_Editing', 'Human_Clinical_Trials', 'Mandatory_Vaccination', 'Organ_Transplant']

min_record = None
min_length = float('inf')
min_topic = ''

for item in file_names:
    df = pd.read_csv(f'all_data/{item}.csv')
    df['word_count'] = df['text'].str.split().str.len()
    
    topic_min_idx = df['word_count'].idxmin()
    topic_min_length = df.loc[topic_min_idx, 'word_count']
    
    if topic_min_length < min_length:
        min_length = topic_min_length
        min_record = df.loc[topic_min_idx]
        min_topic = item

print(f"最小长度: {min_length} 个单词")
print(f"所属话题: {min_topic}")
print(f"文本内容: {min_record['text']}")
print(f"\n完整记录:\n{min_record}")
```
