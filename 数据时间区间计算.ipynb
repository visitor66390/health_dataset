```python
import json
from datetime import datetime

paths = [
    r'C:\A workplace\9 健康医疗立场检测\数据标注收集\data_collection_filtered_ok_news\Euthanasia\comments.json',
    r'C:\A workplace\9 健康医疗立场检测\数据标注收集\data_collection_filtered_ok_news\Gene_Editing\comments.json',
    r'C:\A workplace\9 健康医疗立场检测\数据标注收集\data_collection_filtered_ok_news\Human_Clinical_Trials\comments.json',
    r'C:\A workplace\9 健康医疗立场检测\数据标注收集\data_collection_filtered_ok_news\Mandatory_Vaccination\comments.json',
    r'C:\A workplace\9 健康医疗立场检测\数据标注收集\data_collection_filtered_ok_news\Organ_Transplant\comments.json',
    r'C:\A workplace\9 健康医疗立场检测\数据标注收集\data_collection_filtered_ok_news\Abortion\comments.json',
]

times = [item['createdUTC'] for path in paths for item in json.load(open(path, 'r', encoding='utf-8')).values()]
print(f"时间区间: {datetime.fromtimestamp(min(times))} 至 {datetime.fromtimestamp(max(times))}")
print(f"跨度: {(max(times) - min(times)) / 86400:.1f} 天")
```
